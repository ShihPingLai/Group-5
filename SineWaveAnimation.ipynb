```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML

# 固定起始語法
figure = plt.figure()
axis = plt.axes(xlim=(0, 10), ylim=(-1, 1))
line, = axis.plot([], [], lw=3)
def init():
    line.set_data([], [])
    return line,

# 動畫函數
def animate(i):
    x = np.linspace(0, 10, 5000)
    y = np.sin(np.linspace(0, 100, 5000)*(np.pi * (x - 0.01 * i)))
    line.set_data(x, y)
    return line,


anim = animation.FuncAnimation(figure, animate, init_func=init, frames=200, interval=15, blit=True)
figure.set_size_inches(12, 9, True)
HTML(anim.to_html5_video())
```
